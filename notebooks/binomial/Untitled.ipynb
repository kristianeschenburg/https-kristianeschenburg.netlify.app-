```python
import numpy as np
import scipy

import matplotlib.pyplot as plt
```

```python
alpha=0.75
size=10
P = scipy.stats.norm(loc=1500, scale=500)
sizes = P.rvs(5000).astype(np.int32)
sizes = sizes[sizes>0]

fig, [ax1, ax2, ax3] = plt.subplots(1,3, figsize=(15, 4))
for mask_fraction in [0.25, 0.5, 0.75]:
    B_i = scipy.stats.binom(n=sizes, p=0.025)
    ifx_size = B_i.rvs()

    B_m = scipy.stats.binom(n=ifx_size, p=mask_fraction)
    masked_res = B_m.rvs()

    ax1.scatter(sizes, ifx_size, alpha=alpha, s=size, label=f'MF: {mask_fraction}')
    ax1.set_xlabel('Graph Size')
    ax1.set_ylabel('# Interface residues')

    ax2.scatter(sizes, masked_res, alpha=alpha, s=size, label=f'MF: {mask_fraction}')
    ax2.set_xlabel('Graph Size')
    ax2.set_ylabel('# Masked residues')


    ax3.scatter(ifx_size, masked_res, alpha=alpha, s=size, label=f'MF: {mask_fraction}')
    ax3.set_xlabel('# Interface residues')
    ax3.set_ylabel('# Masked residues')
    
for ax in [ax1, ax2, ax3]:
    ax.legend()
```

```python
N = 10**np.arange(1,4)
```

```python
P = np.linspace(0,1,1000)
```

```python
V = np.sqrt(N[:,None]*(P*(1-P)))
for i in range(len(N)):
    plt.plot(V[i], label=f'N: {N[i]}')
plt.legend()
```
